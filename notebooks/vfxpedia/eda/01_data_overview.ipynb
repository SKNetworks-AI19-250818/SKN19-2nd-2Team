## 📊 상관관계 분석 (한글 라벨링 적용)

히트맵의 축 라벨은 `Variable Decoder`를 사용하여 자동으로 한글로 표시됩니다.

### 📈 히트맵 전략:
1. **전체 변수 히트맵**: 라벨 없이 전체 패턴만 표시 (변수가 너무 많음)
2. **Top 변수 히트맵**: 상위 변수만 선택하여 한글 라벨과 함께 표시

> 💡 **Variable Decoder 사용법**은 `02_decoder_test.ipynb`를 참고하세요.


```python
# 컬럼 타입 분류
print("="*60)
print("📂 변수 타입 분류")
print("="*60)

numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

print(f"\n✅ 숫자형 변수: {len(numeric_cols)}개")
print(numeric_cols[:10], "..." if len(numeric_cols) > 10 else "")

print(f"\n✅ 범주형 변수: {len(categorical_cols)}개")
print(categorical_cols)

print(f"\n총 변수 개수: {len(data.columns)}개")
```

```python
# 상관관계 히트맵 (기본 변수명 라벨)
print("🔥 숫자형 변수 상관관계 분석")

# 모든 숫자형 변수의 상관계수 계산
corr_matrix = data[numeric_cols].corr()

# 히트맵 생성 (변수명 그대로)
plt.figure(figsize=(20, 16))
sns.heatmap(
    corr_matrix,
    annot=False,
    cmap='coolwarm',
    center=0,
    vmin=-1, vmax=1,
    square=True,
    linewidths=0.5,
    cbar_kws={'label': '상관계수'}
)

plt.xticks(
    range(len(corr_matrix.columns)),
    corr_matrix.columns,
    rotation=90,
    ha='right',
    fontsize=6
)
plt.yticks(
    range(len(corr_matrix.columns)),
    corr_matrix.columns,
    rotation=0,
    fontsize=6
)

plt.title('전체 숫자형 변수 상관관계 히트맵', fontsize=11)
plt.tight_layout()
plt.show()

# churn과의 상관계수 Top 10
print("\n📊 churn과 상관관계가 높은 변수 Top 10")
churn_corr = corr_matrix['churn'].sort_values(ascending=False)
churn_corr
```

```python
# Cell 08-01: churn과 상관계수 Top 30 (또는 50)
top_n = 50  # 원하시면 50으로 변경

# churn과의 상관관계 (절댓값 기준)
churn_corr_abs = corr_matrix['churn'].abs().sort_values(ascending=False)
top_vars = churn_corr_abs.head(top_n + 1).index.tolist()  # churn 자기 자신 포함

# 상관관계 높은 변수들만 추출
top_corr = corr_matrix.loc[top_vars, top_vars]

# 실제 값 확인
actual_corr = corr_matrix['churn'][top_vars].sort_values(key=abs, ascending=False)

# 히트맵 (팀 통일 스타일 + 한글 라벨링 - 가독성 개선)
fig, ax = plt.subplots(figsize=(22, 20))  # 크기 증가
sns.heatmap(top_corr, 
            annot=False,
            cmap='coolwarm',
            center=0,
            vmin=-1, vmax=1,
            square=True,
            linewidths=0.3,
            cbar_kws={'label': '상관계수', 'shrink': 0.8},
            ax=ax)

# Top 변수들의 한글 라벨 생성 (simple 형식으로 간결하게)
top_korean_labels = get_korean_labels(top_vars, format_type='simple')

# 축 라벨 설정 (폰트 크기와 간격 조정)
ax.set_xticks(np.arange(len(top_korean_labels)) + 0.5)
ax.set_yticks(np.arange(len(top_korean_labels)) + 0.5)
ax.set_xticklabels(top_korean_labels, rotation=90, ha='right', fontsize=11, va='top')
ax.set_yticklabels(top_korean_labels, rotation=0, fontsize=11, va='center')

# 제목
plt.title(f'Churn과 상관관계 높은 Top {top_n} 변수 히트맵 (한글 라벨)\n', 
          fontsize=16, pad=20)

# 레이아웃 조정 (라벨이 잘리지 않도록)
plt.tight_layout(pad=2.0)
plt.subplots_adjust(bottom=0.15, left=0.15)  # 여백 추가
plt.show()

print(f"💡 Tip: 변수명 전체를 보려면 변수명 위에 마우스를 올려보세요!")

# Cell 08-02: 텍스트로 상위 변수 출력
print(f"\n📊 churn과 상관관계 Top {top_n} (절대값 기준)")
print("=" * 80)

rank = 1
for var, corr_val in actual_corr.items():
    if var != 'churn':  # churn 자기 자신 제외
        direction = "📈 양의 상관" if corr_val > 0 else "📉 음의 상관"
        var_label = decoder.get_variable_label(var)
        print(f"{rank:2d}. {var:20s} ({var_label:30s}) | {corr_val:+.4f} {direction}")
        rank += 1
```
