```python
# 라이브러리 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# 설정
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
sns.set_palette("husl")

# 데이터 로드
data = pd.read_csv("../../../data/analy_data.csv")
print("✅ 데이터 로드 완료")
```

```python
# 1. 데이터 기본 정보
print("="*60)
print("📊 데이터 기본 정보")
print("="*60)

print(f"\n전체 행/열 개수: {data.shape}")
print(f"전체 데이터 포인트: {data.shape[0] * data.shape[1]:,}개")
print(f"메모리 사용량: {data.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print("\n" + "="*60)
print("📋 컬럼 정보")
print("="*60)
data.info()

print("\n" + "="*60)
print("📈 기술 통계량")
print("="*60)
print(data.describe())
```

```python
# 결측치 현황
print("="*60)
print("🔍 결측치 현황")
print("="*60)

missing = data.isnull().sum()
missing_pct = (missing / len(data)) * 100

missing_df = pd.DataFrame({
    '결측치 개수': missing,
    '비율(%)': missing_pct
}).sort_values('결측치 개수', ascending=False)

# 결측치가 있는 컬럼만 출력
missing_cols = missing_df[missing_df['결측치 개수'] > 0]

if len(missing_cols) > 0:
    print(f"\n⚠️ 결측치가 있는 컬럼: {len(missing_cols)}개")
    print(missing_cols)
else:
    print("\n✅ 결측치 없음!")
```

```python
# 결측치 시각화
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1) 결측치 히트맵
if missing_cols.shape[0] > 0:
    sns.heatmap(data[missing_cols.index].isnull(), 
                cbar=True, 
                cmap='YlOrRd',
                yticklabels=False,
                ax=axes[0])
    axes[0].set_title('결측치 위치 히트맵', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('컬럼')
    
    # 2) 결측치 비율 막대그래프
    missing_cols['비율(%)'].plot(kind='barh', ax=axes[1], color='coral')
    axes[1].set_title('결측치 비율 (%)', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('비율 (%)')
else:
    axes[0].text(0.5, 0.5, '결측치 없음', ha='center', va='center', fontsize=20)
    axes[1].text(0.5, 0.5, '결측치 없음', ha='center', va='center', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
# 타겟 분포
print("="*60)
print("🎯 타겟 변수(churn) 분포")
print("="*60)

churn_counts = data['churn'].value_counts().sort_index()
churn_pct = data['churn'].value_counts(normalize=True).sort_index() * 100

result_df = pd.DataFrame({
    '개수': churn_counts,
    '비율(%)': churn_pct
})

print(result_df)
print(f"\n불균형 비율: {churn_counts.max() / churn_counts.min():.2f}:1")

# 해석
if churn_counts.max() / churn_counts.min() > 2:
    print("⚠️ 클래스 불균형 존재 → 모델링 시 주의 필요")
else:
    print("✅ 클래스 균형 양호")
```

```python
# 타겟 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1) 막대그래프
churn_counts.plot(kind='bar', ax=axes[0], color=['#FF6B6B', '#4ECDC4'])
axes[0].set_title('금연 성공/실패 분포', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Churn (0: 실패, 1: 성공)')
axes[0].set_ylabel('개수')
axes[0].set_xticklabels(['실패 (이탈)', '성공'], rotation=0)

# 각 막대 위에 개수 표시
for i, v in enumerate(churn_counts):
    axes[0].text(i, v + 100, f'{v:,}개\n({churn_pct[i]:.1f}%)', 
                 ha='center', fontweight='bold')

# 2) 원그래프
axes[1].pie(churn_counts, 
            labels=['실패 (이탈)', '성공'], 
            autopct='%1.1f%%',
            startangle=90,
            colors=['#FF6B6B', '#4ECDC4'],
            explode=(0.05, 0))
axes[1].set_title('금연 성공률', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 컬럼 타입 분류
print("="*60)
print("📂 변수 타입 분류")
print("="*60)

numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

print(f"\n✅ 숫자형 변수: {len(numeric_cols)}개")
print(numeric_cols[:10], "..." if len(numeric_cols) > 10 else "")

print(f"\n✅ 범주형 변수: {len(categorical_cols)}개")
print(categorical_cols)

print(f"\n총 변수 개수: {len(data.columns)}개")
```

```python
# 상관관계 히트맵 (너무 많으면 샘플링)
print("🔥 숫자형 변수 상관관계 분석")

# 타겟 포함 주요 변수만 선택 (전체는 너무 복잡)
# 일단 모든 숫자형 변수로 시작
corr_matrix = data[numeric_cols].corr()

# 시각화
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, 
            annot=False,  # 너무 많으면 False
            cmap='coolwarm',
            center=0,
            vmin=-1, vmax=1,
            square=True,
            linewidths=0.5)
plt.title('전체 숫자형 변수 상관관계 히트맵', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# churn과의 상관계수 Top 10
print("\n📊 churn과 상관관계가 높은 변수 Top 10")
churn_corr = corr_matrix['churn'].sort_values(ascending=False)
# print(churn_corr.head(11))  # churn 자기 자신 포함이라 11개
churn_corr
```

```python
# Cell 08-01: churn과 상관계수 Top 30 (또는 50)
top_n = 50  # 원하시면 50으로 변경

# churn과의 상관관계 (절댓값 기준)
churn_corr_abs = corr_matrix['churn'].abs().sort_values(ascending=False)
top_vars = churn_corr_abs.head(top_n + 1).index.tolist()  # churn 자기 자신 포함

# 상관관계 높은 변수들만 추출
top_corr = corr_matrix.loc[top_vars, top_vars]

# 실제 값 확인
actual_corr = corr_matrix['churn'][top_vars].sort_values(key=abs, ascending=False)

# 히트맵
plt.figure(figsize=(18, 16))
sns.heatmap(top_corr, 
            annot=False,  # 30개면 숫자 안 넣는 게 나음
            cmap='coolwarm',
            center=0,
            vmin=-1, vmax=1,
            square=True,
            linewidths=0.5,
            cbar_kws={'label': '상관계수'})
plt.title(f'Churn과 상관관계 높은 Top {top_n} 변수 히트맵', 
          fontsize=16, fontweight='bold')
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Cell 08-02: 텍스트로 상위 변수 출력
print(f"\n📊 churn과 상관관계 Top {top_n} (절대값 기준)")
print("=" * 60)

rank = 1
for var, corr_val in actual_corr.items():
    if var != 'churn':  # churn 자기 자신 제외
        direction = "📈 양의 상관" if corr_val > 0 else "📉 음의 상관"
        print(f"{rank:2d}. {var:20s} | {corr_val:+.4f} {direction}")
        rank += 1
```

```python
# Cell 08-07: 변수명 디코딩 및 해석

# 변수명 매핑 딕셔너리
var_dict = {
    'age': {
        'name': '만 나이',
        'values': None  # 연속형
    },
    'nua_01z2': {
        'name': '아침식사 빈도',
        'values': {
            1: '주 5~7회',
            2: '주 3~4회',
            3: '주 1~2회',
            4: '거의 안함(주 0회)'
        }
    },
    'sod_02z3': {
        'name': '혼인상태',
        'values': {
            1: '배우자와 동거',
            2: '배우자와 별거',
            3: '사별',
            4: '이혼',
            5: '미혼'
        }
    },
    'drb_01z3': {
        'name': '연간 음주 빈도',
        'values': {
            1: '최근 1년간 전혀 마시지 않았다',
            2: '월 1회 미만',
            3: '월 1회 정도',
            4: '월 2~4회',
            5: '주 2~3회',
            6: '주 4회 이상',
            8: '비해당',
            9: '모름'
        }
    },
    'sob_01z1': {
        'name': '교육수준(최종학력)',
        'values': {
            1: '무학',
            2: '서당/한학',
            3: '초등학교',
            4: '중학교',
            5: '고등학교',
            6: '2-3년제 대학',
            7: '4년제 대학',
            8: '대학원 이상'
        }
    },
    'soa_01z1': {
        'name': '경제활동 여부',
        'values': {
            1: '예 (최근 1주일간 수입목적 근로)',
            2: '아니오'
        }
    }
}

def decode_var_output(var_name, grouped_df):
    """변수명 디코딩하여 출력"""
    info = var_dict.get(var_name, {'name': var_name, 'values': None})
    
    print(f"\n📊 {var_name} ({info['name']}):")
    
    if info['values']:
        # 값 디코딩
        decoded_df = grouped_df.copy()
        decoded_df.index = decoded_df.index.map(
            lambda x: f"{x} ({info['values'].get(x, '알 수 없음')})"
        )
        print(decoded_df.to_string())
    else:
        print(grouped_df.to_string())

# 실행
print("🎯 핵심 변수별 금연 성공률 (디코딩)")
print("="*60)

for var in ['age', 'nua_01z2', 'sod_02z3', 'drb_01z3', 'sob_01z1', 'soa_01z1']:
    grouped = data.groupby(var)['churn'].agg(['mean', 'count'])
    grouped['성공률(%)'] = grouped['mean'] * 100
    grouped = grouped[grouped['count'] > 100]
    result = grouped[['성공률(%)', 'count']].sort_values('성공률(%)', ascending=False)
    
    decode_var_output(var, result)
```
