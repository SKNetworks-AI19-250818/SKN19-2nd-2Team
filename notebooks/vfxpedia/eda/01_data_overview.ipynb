```python
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ (Í∏∞Î≥∏ Î≥ÄÏàòÎ™Ö ÎùºÎ≤®)
print("üî• Ïà´ÏûêÌòï Î≥ÄÏàò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Î∂ÑÏÑù")

# Î™®Îì† Ïà´ÏûêÌòï Î≥ÄÏàòÏùò ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
corr_matrix = data[numeric_cols].corr()

# ÌûàÌä∏Îßµ ÏÉùÏÑ± (Î≥ÄÏàòÎ™Ö Í∑∏ÎåÄÎ°ú)
plt.figure(figsize=(20, 16))
sns.heatmap(
    corr_matrix,
    annot=False,
    cmap='coolwarm',
    center=0,
    vmin=-1, vmax=1,
    square=True,
    linewidths=0.5,
    cbar_kws={'label': 'ÏÉÅÍ¥ÄÍ≥ÑÏàò'}
)

plt.xticks(
    range(len(corr_matrix.columns)),
    corr_matrix.columns,
    rotation=90,
    ha='right',
    fontsize=6
)
plt.yticks(
    range(len(corr_matrix.columns)),
    corr_matrix.columns,
    rotation=0,
    fontsize=6
)

plt.title('Ï†ÑÏ≤¥ Ïà´ÏûêÌòï Î≥ÄÏàò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ', fontsize=11)
plt.tight_layout()
plt.show()

# churnÍ≥ºÏùò ÏÉÅÍ¥ÄÍ≥ÑÏàò Top 10
print("\nüìä churnÍ≥º ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÜíÏùÄ Î≥ÄÏàò Top 10")
churn_corr = corr_matrix['churn'].sort_values(ascending=False)
churn_corr
```

```python
# Cell 08-01: churnÍ≥º ÏÉÅÍ¥ÄÍ≥ÑÏàò Top 30 (ÎòêÎäî 50)
top_n = 50  # ÏõêÌïòÏãúÎ©¥ 50ÏúºÎ°ú Î≥ÄÍ≤Ω

# churnÍ≥ºÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ (Ï†àÎåìÍ∞í Í∏∞Ï§Ä)
churn_corr_abs = corr_matrix['churn'].abs().sort_values(ascending=False)
top_vars = churn_corr_abs.head(top_n + 1).index.tolist()  # churn ÏûêÍ∏∞ ÏûêÏã† Ìè¨Ìï®

# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÎÜíÏùÄ Î≥ÄÏàòÎì§Îßå Ï∂îÏ∂ú
top_corr = corr_matrix.loc[top_vars, top_vars]

# Ïã§Ï†ú Í∞í ÌôïÏù∏
actual_corr = corr_matrix['churn'][top_vars].sort_values(key=abs, ascending=False)

# ÌûàÌä∏Îßµ (ÌåÄ ÌÜµÏùº Ïä§ÌÉÄÏùº + ÌïúÍ∏Ä ÎùºÎ≤®ÎßÅ - Í∞ÄÎèÖÏÑ± Í∞úÏÑ†)
fig, ax = plt.subplots(figsize=(22, 20))  # ÌÅ¨Í∏∞ Ï¶ùÍ∞Ä
sns.heatmap(top_corr, 
            annot=False,
            cmap='coolwarm',
            center=0,
            vmin=-1, vmax=1,
            square=True,
            linewidths=0.3,
            cbar_kws={'label': 'ÏÉÅÍ¥ÄÍ≥ÑÏàò', 'shrink': 0.8},
            ax=ax)

# Top Î≥ÄÏàòÎì§Ïùò ÌïúÍ∏Ä ÎùºÎ≤® ÏÉùÏÑ± (simple ÌòïÏãùÏúºÎ°ú Í∞ÑÍ≤∞ÌïòÍ≤å)
top_korean_labels = get_korean_labels(top_vars, format_type='simple')

# Ï∂ï ÎùºÎ≤® ÏÑ§Ï†ï (Ìè∞Ìä∏ ÌÅ¨Í∏∞ÏôÄ Í∞ÑÍ≤© Ï°∞Ï†ï)
ax.set_xticks(np.arange(len(top_korean_labels)) + 0.5)
ax.set_yticks(np.arange(len(top_korean_labels)) + 0.5)
ax.set_xticklabels(top_korean_labels, rotation=90, ha='right', fontsize=11, va='top')
ax.set_yticklabels(top_korean_labels, rotation=0, fontsize=11, va='center')

# Ï†úÎ™©
plt.title(f'ChurnÍ≥º ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÎÜíÏùÄ Top {top_n} Î≥ÄÏàò ÌûàÌä∏Îßµ (ÌïúÍ∏Ä ÎùºÎ≤®)\n', 
          fontsize=16, pad=20)

# Î†àÏù¥ÏïÑÏõÉ Ï°∞Ï†ï (ÎùºÎ≤®Ïù¥ ÏûòÎ¶¨ÏßÄ ÏïäÎèÑÎ°ù)
plt.tight_layout(pad=2.0)
plt.subplots_adjust(bottom=0.15, left=0.15)  # Ïó¨Î∞± Ï∂îÍ∞Ä
plt.show()

print(f"üí° Tip: Î≥ÄÏàòÎ™Ö Ï†ÑÏ≤¥Î•º Î≥¥Î†§Î©¥ Î≥ÄÏàòÎ™Ö ÏúÑÏóê ÎßàÏö∞Ïä§Î•º Ïò¨Î†§Î≥¥ÏÑ∏Ïöî!")

# Cell 08-02: ÌÖçÏä§Ìä∏Î°ú ÏÉÅÏúÑ Î≥ÄÏàò Ï∂úÎ†•
print(f"\nüìä churnÍ≥º ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Top {top_n} (Ï†àÎåÄÍ∞í Í∏∞Ï§Ä)")
print("=" * 80)

rank = 1
for var, corr_val in actual_corr.items():
    if var != 'churn':  # churn ÏûêÍ∏∞ ÏûêÏã† Ï†úÏô∏
        direction = "üìà ÏñëÏùò ÏÉÅÍ¥Ä" if corr_val > 0 else "üìâ ÏùåÏùò ÏÉÅÍ¥Ä"
        var_label = decoder.get_variable_label(var)
        print(f"{rank:2d}. {var:20s} ({var_label:30s}) | {corr_val:+.4f} {direction}")
        rank += 1
```
